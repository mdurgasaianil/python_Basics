### 10.

The break statement terminates the whole iteration of a loop whereas continue skips the current iteration.

EX:
```python
a = 4
while a > 0:
    if a == 2:
        print(a)
        break  # if this condition executes it will terminate the loop and execute the outside loop print statement
    if a == 4:
        print(a)
        a -=1
        continue # if this condition executes it skips the below print and airthmetic operation in that iteration
    print("Not used continue")
    a -=1
print("loop Terminated at",a)
```
OUTPUT:<br>
4<br>
Not used continue<br>
2<br>
loop Terminated at 2
