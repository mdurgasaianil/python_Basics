## Questions

1. Write a Python program to find words which are greater than given length k?
2. Write a Python program for removing i-th character from a string?
3. Write a Python program to split and join a string?
4. Write a Python to check if a given string is binary string or not?
5. Write a Python program to find uncommon words from two Strings?
6. Write a Python to find all duplicate characters in string?
7. Write a Python Program to check if a string contains any special character?

## Answers

### 1.

```python
def findWords(words,length):
    word = ""
    for i in words:
        if len(i) >= length:
            word += " "+i
    return word.lstrip() if word else "No words are greater than given length {}".format(length)
words = ["anil","dinesh","avez","phani","harsha"]
findWords(words,6)
```

### 2.

```python
def removeIthChar(string,pos):
    s = ""
    for n,i in enumerate(string):
        if n != pos:
            s += i
    return s
removeIthChar("Anil",3)
```

### 3.

```python
def splitAndJoin(string,spli,joi):
    return joi.join(string.split(spli))
splitAndJoin("Hello,My,world",","," ")
```

### 4.

```python
def checkBinary(stri):
    try:
        stri = int(stri)
        for i in str(stri):
            if i != "0" and i != "1":
                return "Not Binary String"
        return "Binary String"
    except:
        return "Not Binary String"
checkBinary("1001010")
```

### 5.

```python
def uncommon(string1,string2):
    s1 = [s1 for s1 in string1.split(" ")]
    s2 = [s2 for s2 in string2.split(" ")]
    unwords1 = [w1 if w1 not in s2 else "" for w1 in s1]
    unwords2 = [w2 if w2 not in s1 else "" for w2 in s2]
    return list(filter(None,unwords1 + unwords2))
string1 = "India is my country"
string2 = "i am an indian country citizen"
uncommon(string1,string2)
```

### 6.

```python
def countOccurences(li):
    from collections import defaultdict
    c = defaultdict(int)
    for i in li:
        if c[i]:
            c[i] += 1
        else:
            c[i] = 1
    return c
def duplicate(string):
    h = ""
    for k,v in countOccurences(string).items():
        h += k if v > 1 else ""
    return h
duplicate("abcssdefgghhh")
```

### 7.
```python
def checkSC(string):
    SC = ["@","_","!","#","$","%","^","&","*","(",")","<",">","?","/","'\'","|","{","}","~",":","[","]"]
    for i in string:
        if i in SC:
            return "Special Character is there"
    return "No Special Character"
checkSC("[Anil")
```
