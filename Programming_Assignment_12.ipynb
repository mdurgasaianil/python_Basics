### 1.
```python
def uniqueDictValues(dictionary):
    return list({val for val in dictionary.values()})
dic = {"anil":"Associate",
      "Dinesh":"Student",
      "Avez":"SDE",
      "Harshi":"SDE",
      "Bharani":"Associate",
      "RRG":"Student"}      
uniqueDictValues(dic)
```

### 2.
```python
dic = {"anil":48000,
      "Dinesh":1000,
      "Avez":55000,
      "Harshi":55000,
      "Bharani":48000,
      "RRG":1000}
def sumOfValues(dic):
    return sum(dic.values())
sumOfValues(dic)
```

```python
dic1 = {"anil":48000,
      "Dinesh":1000,
      "Avez":55000,
      "Harshi":55000,
      "Bharani":48000,
      "RRG":1000}
dic2 = {"anil1":"Associate",
      "Dinesh2":"Student",
      "Avez3":"SDE",
      "Harshi4":"SDE",
      "Bharani5":"Associate",
      "RRG6":"Student"}
def mergeDict(dic1,dic2):
    dic1.update(dic2)
    return dic1   
mergeDict(dic1,dic2)
```

### 6.
```python
def checkOrder(string,pattern):
    from collections import OrderedDict
    dic = OrderedDict.fromkeys(string)
    patternLen = 0
    for key in dic.keys():
        if key == pattern[patternLen]:
            patternLen +=1
        if patternLen == len(pattern):
            return "Order Matched"
    return "Order Mismatched"
checkOrder("anil","la")
```
