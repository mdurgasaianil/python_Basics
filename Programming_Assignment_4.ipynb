## Answers

### 1.

```python
def fact(num):
    val = 1
    for i in range(1,num+1):
        val = val * i
    return val
fact(5)
```

### 2.

```python
def mulTable(num):
    for i in range(1,11):
        print(num,"*",i,"=",int(num*i))
mulTable(3)
```

### 3.

```python
def fibinacci(num):
    seq = [0,1]
    ind1 = 0
    ind2 = 1
    while num:
        seq.append(seq[ind1] + seq[ind2])
        ind1 += 1
        ind2 += 1
        num -= 1
    return seq[2:]
fibinacci(10)
```

### 4.

```python
def checkArmstrong(num):
    val = num
    sol = []
    while num:
        sol.append((num%10)**3)
        num = num//10
    if val == sum(sol):
        return "Armstrong Number"
    else:
        return "Not a Armstrong Number"
checkArmstrong(372)
```

### 5.

```python
def checkArmstrong(num):
    val = num
    sol = []
    while num:
        sol.append((num%10)**3)
        num = num//10
    if val == sum(sol):
        return "Armstrong Number"
    else:
        return "Not a Armstrong Number"
def printArmstrong(start,stop):
    values = []
    for i in range(start,stop):
        if i > 10:
            if checkArmstrong(i) == "Armstrong Number":
                values.append(i)
    if values:
        return values
    else:
        return "No Armstrong Number is there in this interval"
printArmstrong(100,200)
```
### 6.

```python
def numCheck(num):
    if num > 0:
        return "Positive number"
    elif num == 0:
        return "Zero"
    else:
        return "Negative Number"
def sumOfNNo(nums):
    s = 0
    for i in nums:
        if numCheck(i) != "Negative Number" and numCheck(i) != "Zero":
            s += i
    return s
sumOfNNo([0,1,2,4,-9])
```
