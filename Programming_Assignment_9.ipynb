## Answers

### 1.

```python
def diasariumCheck(num):
    s = 0
    c = 1
    for i in str(num):
        s += int(i)**c
        c += 1
    if s == num:
        return "Disarium Number"
    else:
        return "Not a Disarium Number"
```

### 2.

```python
def diasariumCheck(num):
    s = 0
    c = 1
    for i in str(num):
        s += int(i)**c
        c += 1
    if s == num:
        return "True"
    else:
        return "False"
def printDiasarium(num_range):
    nums = ""
    for i in range(1,num_range):
        if diasariumCheck(i) == "True":
            nums += " "+str(i)
    return nums
printDiasarium(100)
```

### 3.

```python
def happyNumCheck(num):
    if len(str(num)) == 1:
        if num == 1:
            return "Happy Number"
        else:
            val = num**2
            return happyNumCheck(val)
    else:
        s = 0
        for i in str(num):
            s += int(i)**2
        if s < 10 and s == 1:
            return happyNumCheck(s)
        elif s >= 10:
            return happyNumCheck(s)
        else:
            return "Not Happy Number"
happyNumCheck(19)
```

### 4.

```python
def happyNumCheck(num):
    if len(str(num)) == 1:
        if num == 1:
            return True
        else:
            val = num**2
            return happyNumCheck(val)
    else:
        s = 0
        for i in str(num):
            s += int(i)**2
        if s < 10 and s == 1:
            return happyNumCheck(s)
        elif s >= 10:
            return happyNumCheck(s)
        else:
            return False
def printHappyNums(num_range):
    val = ""
    for i in range(1,num_range):
        if happyNumCheck(i):
            val += " "+str(i)
    return val
printHappyNums(100)

```

### 6.

```python
def harshadsCheck(num):
    s = 0
    for i in str(num): s += int(i)
    return True if num%s == 0 else False
harshadsCheck(157)
def printHarshads(num_range):
    val = ""
    for i in range(1,num_range):
        val += " "+str(i) if harshadsCheck(i) else ""
    return val
printHarshads(100)
```
