def sumofele(li):
    s = 0
    if len(li)==1:
        return li[0]  
    else:
        s += li.pop()
        s += sumofele(li)
        return s

### 5.

```python
a = [54,23,51,9,24,3,6,4,2,11]
def secondLargest(li):
    t = li[:]
    t.remove(max(t))
    return max(t)
print(a,secondLargest(a))
```

### 6.

```python
def nlargest(li,n):
    a = ""
    t = li[:]
    for _ in range(n):
        a += " "+str(max(t))
        t.remove(max(t))
    return a.lstrip(' ')
import random
res = random.sample(range(1, 1000), 500)
nlargest(res,100)
```

### 7.
```python
def even(no_range):
    return list(i for i in range(no_range) if i%2 == 0)
print(even(100))
```

### 8.

```python
def odd(no_range):
    return list(i for i in range(no_range) if i%2 != 0)
print(odd(100))
```

### 11.

```python
from collections import defaultdict
a = [54,23,51,9,24,3,6,4,2,11,3,5,5,5,"ada","anil","fds"]
def countOccurences(li):
    c = defaultdict(int)
    for i in li:
        if c[i]:
            c[i] += 1
        else:
            c[i] = 1
    return c
for k,v in countOccurences(a).items():
    print(k,":",v)
```
