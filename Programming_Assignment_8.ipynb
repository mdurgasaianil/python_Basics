## Answers

### 1. 

```python
def displayMatrix(a,b):
    for i,j in zip(a,b):
        print(i,j)
def addMatrices(mat1,mat2):
    c = []
    for i in range(len(mat1)):
        d = []
        for j in range(len(mat1[i])):
            d.append(mat1[i][j]+mat2[i][j])
        c.append(d)
    displayMatrix(mat1,mat2)
    return c
a,b = [[1,2,3],[4,5,6],[7,8,9]],[[7,8,9],[4,5,6],[1,2,3]]
addMatrices(a,b)
```

### 2.

```python
a,b = [[1,2,3],[4,5,6],[7,8,9],[250,4,1]],[[7,8,1,8],[4,5,2,1],[1,2,8,6]]
def matrix(a,b):
    if len(a[0]) == len(b):
        c = []
        for i in range(len(a)):
            d = []
            for j in range(len(b[0])):
                s = 0
                for k in range(len(a[0])):
                    s += a[i][k]*b[k][j]
                d.append(s)
            c.append(d)
        return c
    else:
        return "Matrices cannot multiplicated"
matrix(a,b)
```

### 3.

```python
mat = [[1,2],[3,4],[5,6]]
def Transpose(mat):
    transpose = []
    for i in range(len(mat[0])):
        temp = []
        for j in range(len(mat)):
            temp.append(mat[j][i])
        transpose.append(temp)
    return transpose
print(mat)
Transpose(mat)
```

### 5.

```python
def removePunct(string):
    a = ['!',"@","#","$","%","^","-","_",".",",","/","'",'"',";",":","|"]
    for i in a:
        string = ''.join(string.split(i))
    return string
removePunct("Anil',s it's been")
```
